```python
subset = df[["total_bill","tip","day"]]
subset.head()
```

```python
#filter bill greater than 25
df[df["total_bill"] > 25]
```

```python
#filter smokers only
df[df["smoker"] == "Yes"]
```

```python
#filter female customers at dinner
df[(df["sex"] == "Female") & (df["time"] == "Dinner")]
```

```python
#create new column tip percentage
df["tip_percentage"] = df["tip"] / df["total_bill"]
df.head(10)
```

```python
#find missing values
df.isnull().sum()
```

```python
#save cleaned data set
df.to_csv("day1_clean.csv",index=False)
```
